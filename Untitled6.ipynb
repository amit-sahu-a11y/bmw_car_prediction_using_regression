```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/bmw.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
x=df.drop(columns=['price'],axis=1)
y=df['price']
```

```python
meann=df['price'].mean()
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
df.head()
```

```python
df['fuelType'].value_counts()
```

```python
cat=['model','transmission','fuelType']
num=['year','mileage','tax','mpg','engineSize']
```

```python
cat_pipe=Pipeline(
    steps=[
        ("encoded",OneHotEncoder(drop="first",sparse_output=False,handle_unknown='ignore')),
    ]
)
```

```python
num_pip=Pipeline(
    steps=[
        ("scaled",StandardScaler())
    ]
)
```

```python
preprocessor=ColumnTransformer(
    transformers=[
        ('cat',cat_pipe,cat),
        ('num',num_pip,num)
    ]
)
```

```python
final=Pipeline(
    steps=[
        ('preprocessor',preprocessor),
        ('model',LinearRegression())
    ]
)
```

```python
final.fit(x_train,y_train)
```

```python
x_train.head()
```

```python
from sklearn.metrics import mean_absolute_error
y_train_pred=final.predict(x_train)
error=mean_absolute_error(y_train,y_train_pred)
error
```

```python
er=(error/y_train.mean())*100
er
```

```python
from sklearn.metrics import r2_score

y_pred = final.predict(x_train)
print(r2_score(y_train, y_pred))
```

```python
from sklearn.metrics import r2_score

y_pred = final.predict(x_test)
print(r2_score(y_test, y_pred))
```
